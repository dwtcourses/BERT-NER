```python
import os
import zipfile

def zipdir(path, ziph):
    # ziph is zipfile handle
    for root, dirs, files in os.walk(path):
        for file in files:
            ziph.write(os.path.join(root, file))
  
zipf = zipfile.ZipFile('bert_out.zip', 'w', zipfile.ZIP_DEFLATED)
zipdir('out_bert_large', zipf)
zipf.close()
```
